```python
import random
import math
import statistics as stat
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def coin_flip():
    """Let 1 signify heads and 0 signify tails."""
    return random.randint(0, 1)
```

```python
def powers_of_two_with_n(n=100000):
    """Returns a list of the powers of two less than n, and n."""
    powers_of_two = []
    under_n = True
    current_exponent = 0
    while under_n:
        if 2**current_exponent > n:
            under_n = False
        else: 
            powers_of_two.append(2**current_exponent)
            current_exponent += 1
    powers_of_two.append(n)
    return powers_of_two


def coin_trial(num_trials=100000):
    """Runs a coin-flipping simulation for num_trials rounds, where num_trials defaults to 100000"""
    # precompute powers of two less than num_trials.
    powers_of_two = []
    under_n = True
    current_exponent = 0
    while under_n:
        if 2**current_exponent > num_trials:
            under_n = False
        else: 
            powers_of_two.append(2**current_exponent)
            current_exponent += 1
    trial_list = []
    heads_total = 0
    for trial_num in range(1, num_trials + 1):
        heads_total += coin_flip()
        if trial_num in powers_of_two:
            trial_list.append((trial_num, heads_total))
    trial_list.append((num_trials, heads_total))
    return trial_list
```

```python
trial_1 = coin_trial()

def string_power_of_two(given_num):
    power = math.log2(given_num)
    exponential_string = "2**{}".format(round(power, 2))
    return exponential_string

def diff_heads_minus_tails(given_tuple):
    return given_tuple[1] - given_tuple[0] + given_tuple[1]


def ratio_heads_tails(given_tuple):
    if (given_tuple[0] - given_tuple[1]) is 0:
        return None
    return given_tuple[1] / (given_tuple[0] - given_tuple[1])


print(trial_1)

diff_heads = [diff_heads_minus_tails(heads) for heads in trial_1]
print(diff_heads)

ratio_heads = [ratio_heads_tails(heads) for heads in trial_1]
print(ratio_heads)
```

```python
trial_list = []
for _ in range(20):
    trial_list.append(coin_trial())
ratio_list = []
for _ in range(20):
    pre_ratio_trial = coin_trial()
    ratio_trial = [(trial[0], ratio_heads_tails(trial))
                   for trial in pre_ratio_trial]
    ratio_list.append(ratio_trial)
print(ratio_list)
```
